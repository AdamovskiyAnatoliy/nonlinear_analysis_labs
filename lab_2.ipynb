# Лабораторна робота №2

## Варіант №1
$$с = -0,11 + 0,51i$$

```python
def show_fractal(N, xmin, xmax, ymin, ymax, title, cmap='hot'):
    plt.figure(figsize=(10, 10))
    plt.title(title)
    plt.imshow(N, cmap=cmap, extent=[xmin, xmax, ymin, ymax]);
```

## Факторіал Мандельброта

```python
def mandelbrot_set(xmin=-2, xmax=1, 
                   ymin=-1.5, ymax=1.5,
                   xnum=1000, ynum=1000,
                   maxiter=100, a=1, b=0, c=1, 
                   threshold=4, degree=2,
                   cmap='hot'):
    X = np.linspace(xmin, xmax, xnum)
    Y = np.linspace(ymin, ymax, ynum)
    C = a * X + complex(b, c) * Y[:, np.newaxis]
    Z = np.zeros_like(C)
    N = np.zeros_like(C, dtype=int)
    for n in tqdm_notebook(range(maxiter)):
        I = np.less(np.abs(Z), threshold)
        N[I] = n
        Z[I] = Z[I]**degree + C[I]
    show_fractal(N, xmin, xmax, ymin, ymax, title='Фрактал Мандельброта', cmap=cmap)
    return Z, N
```

```python
Z, N = mandelbrot_set()
```

```python
Z, N = mandelbrot_set(maxiter=300, ymax=0.05, ymin=-0.05, xmin=-1.8, xmax=-1.74)
```

## Факторіал Жюліа

```python
def julia_set(xmin=-1.5, xmax=1.5, 
              ymin=-1.5, ymax=1.5,
              xnum=1000, ynum=1000,
              maxiter=100, a=1, b=0, c=1, 
              с_a=-0.5, c_b=0.5,
              threshold=10**6, degree=2):
    X = np.linspace(xmin, xmax, xnum)
    Y = np.linspace(ymin, ymax, ynum)
    C = complex(с_a, c_b)
    Z = a * X + complex(b, c) * Y[:, np.newaxis]
    N = np.zeros_like(Z, dtype=int)
    for n in tqdm_notebook(range(maxiter)):
        I = np.less(np.abs(Z), threshold)
        N[I] = n
        Z[I] = Z[I]**degree + C
    show_fractal(N, xmin, xmax, ymin, ymax, title='Фрактал Жюліа')
    return Z, N
```

```python
# с_a = 0.1
# c_b = 0.7
с_a = -0.11
c_b = 0.51
c_a = np.sqrt(3) / 2
c_b = np.sqrt(1) / 2
Z, N = julia_set(maxiter=100, с_a=с_a, c_b=c_b)
```
